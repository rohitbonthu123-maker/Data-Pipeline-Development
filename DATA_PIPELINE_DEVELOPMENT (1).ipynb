```python
# ==========================================
# ETL Data Pipeline in Google Colab
# ==========================================

# Install required libraries (if needed)
!pip -q install pandas scikit-learn openpyxl

# Import libraries
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, StandardScaler
from google.colab import files

# ------------------------------------------
# STEP 1: Upload Dataset
# ------------------------------------------
print("Please upload a CSV or Excel file")
uploaded = files.upload()

# Get uploaded file name
filename = list(uploaded.keys())[0]

# Read dataset based on file type
if filename.endswith('.csv'):
    df = pd.read_csv(filename)
elif filename.endswith('.xlsx') or filename.endswith('.xls'):
    df = pd.read_excel(filename)
else:
    raise ValueError("Unsupported file format. Please upload a CSV or Excel file.")

print("\nOriginal Dataset Shape:", df.shape)
print(df.head())

# ------------------------------------------
# STEP 2: Remove Duplicate Rows
# ------------------------------------------
duplicates = df.duplicated().sum()
df.drop_duplicates(inplace=True)

print("\nDuplicate Rows Removed:", duplicates)

# ------------------------------------------
# STEP 3: Handle Missing Values
# ------------------------------------------

# Numerical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

if len(num_cols) > 0:
    num_imputer = SimpleImputer(strategy='mean')
    df[num_cols] = num_imputer.fit_transform(df[num_cols])

# Categorical columns
cat_cols = df.select_dtypes(include=['object']).columns

for col in cat_cols:
    # Fix for FutureWarning: Assign the result of fillna back to the column
    df[col] = df[col].fillna(df[col].mode()[0])

print("\nMissing Values Handled Successfully")

# ------------------------------------------
# STEP 4: Encode Categorical
```
